# Distinguishing Hate Speech from Offensive Language on Twitter Using Machine Learning Techniques

## Why This Matters
Every second, thousands of tweets are posted. Some contain harmful language that can impact individuals and communities.

This project explores how machine learning can automatically detect and distinguish:
- Hate Speech (targeted harm)
- Offensive Language (abusive but not hate)
- Neutral content

## Key Challenge
The biggest challenge is:
- **Class imbalance**
- **Subtle differences between hate and offensive language**

The project uses the Davidson et al. (2017) Hate Speech and Offensive Language dataset and evaluates different machine learning models for multi-class text classification.

## Target Variable
The target variable is **`class`**, which is mapped as:
- 0 → Hate Speech  
- 1 → Offensive Language  
- 2 → Neither

## Research Questions

1. Can machine learning models accurately distinguish hate speech from offensive language on Twitter?
2. How does dataset imbalance affect classification performance?
3. Does balancing the dataset improve the performance of machine learning models?
4. Which machine learning model performs best for hate speech detection?

## Dataset Description

The dataset used in this project is the Hate Speech and Offensive Language dataset created by Davidson et al. (2017).

Dataset Characteristics:
- Source: Twitter
- Total tweets: ~24,783
- Annotation method: Crowd-sourced labeling

Class Labels:
0 → Hate Speech

1 → Offensive Language

2 → Neither

Each tweet was annotated by multiple crowd workers to determine whether it contains hate speech, offensive language, or neither.
